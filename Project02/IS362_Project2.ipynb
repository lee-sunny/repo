## For my analysis, I will attempt to determine which has the highest hiring status based on the month where the Job Application was submitted.

### Loading pandas / numpy / datetime modules in Jupyter notebook

```python
import numpy as np
import pandas as pd
import datetime
```

### Loading dataset into Jupyter notebook

```python
jobapp = pd.read_csv(r'C:\Users\Daddy\Videos\IS362_Project2-master\job_app_status.csv')
```

### Looking at the top 5 records of the dataset

As you can tell, there are (.) dots or periods that are in place of a null value that is needed. Since there is a dot, I believe it is important to check the data types for each column:

### Counting the total records of each data series on this dataset

```python
jobapp.count()
```

### Extracting the month from the Application Date data series

```python
jobapp['month'] = pd.DatetimeIndex(jobapp['Application Date']).month
jobapp.head()
```

### Now we will filtered the 'Status' data series to only show 'Hired' status

```python
onlyhired = jobapp[jobapp.Status=='Hired']
onlyhired.head()
```

### We will use the groupby, count and sort to determine which month has the highest hiring rate based on the month of job application submission.

```python
onlyhired.groupby('month')['App Index'] \
.count() \
.reset_index(name='count') \
.sort_values(['count'], ascending=False)
```

# Based on our analysis of the data set, we have can concluded that the month with the highest hiring rate is May.
